## 2️⃣ Veri Seti Hazırlığı

```python
# Temel veri setini oluştur
from scripts.data_collection import create_base_dataset

print("📊 Veri seti oluşturuluyor...\n")
create_base_dataset()

print("\n✅ Veri seti hazır!")
```

```python
# Veri setini kontrol et
import json
from pathlib import Path

data_dir = Path('dataset/raw')
total_samples = 0

print("📁 Veri Dosyaları:\n")
for json_file in data_dir.glob('*.json'):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        count = len(data)
        total_samples += count
        print(f"  {json_file.name}: {count:,} örnek")

print(f"\n📊 Toplam: {total_samples:,} örnek")
```

```python
# Örnek veri göster
sample_file = list(data_dir.glob('*.json'))[0]
with open(sample_file, 'r', encoding='utf-8') as f:
    data = json.load(f)
    sample = data[0]

print("📝 Örnek Veri:\n")
print(f"Kategori: {sample.get('category', 'N/A')}")
print(f"\nTalimat:\n{sample.get('instruction', 'N/A')}")
print(f"\nYanıt:\n{sample.get('response', 'N/A')[:200]}...")
```
